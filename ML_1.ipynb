```
import tensorflow as tf
from tensorflow import keras

import numpy as np

model = keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])]) # Dense type of layer, only one layer
#dense is a simple type of layer with basic neurons in it.

# One unit = One Neuron


# Keras is a framework within tensor flow
#  for making ml models easy.
#  Sequential means we are going to define a swquence of layers
model.compile(optimizer='sgd',loss='mean_squared_error')

# the core of how network learn is in two functions

# We guess the realtionship between data and 

# optmizer function
# loss function  -  How good or bad the guess is.

# Result is used by another optimizer to make another guess.
# Then again with logic being optmizer can improve the guess.

# loop - Make a guess, measure effectiveness with loss function, optimize to get another guess and repeat 500 times.

# SGD - Stochastic Gradient Descent

xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0],dtype=float)
ys = np.array([-3.0,-1.0,1.0,3.0,5.0,7.0],dtype=float)

# np - numbpy , a python library that makes data handling easier.

model.fit(xs,ys,epochs=500)

#  Learning and do the loop

print(model.predict([10.0]))
```
